## Tasks

Now write and execute SQL queries to solve the assignment tasks.

### Task 1




##### Display the names of the unique launch sites  in the space mission


```python
#%sql select DISTINCT LAUNCH_SITE from SPACEXDATASET

df_cleaned['Launch_Site'].unique()
```


### Task 2


#####  Display 5 records where launch sites begin with the string 'CCA' 


```python
#%sql select * from SPACEX where LaunchSite like 'CCA%" limit 5"

df[df['Launch_Site'].str.startswith('CCA')].head()
```

### Task 3




##### Display the total payload mass carried by boosters launched by NASA (CRS)


```python
#%sql select sum(PAYLOAD_MASS_KG_) as Total_PayloadMassKG from SPACEX where Customer like 'NASA (CRS)'

total_payload_mass_nasa_crs = df.loc[df['Customer'] == 'NASA (CRS)', 'PAYLOAD_MASS__KG_'].sum()

total_payload_mass_nasa_crs
```

### Task 4




##### Display average payload mass carried by booster version F9 v1.1


```python
#%sql select AVG(PAYLOAD_MASS_KG_) as Avg_PayloadMass from SPACEX where BoosterVersion = 'F9 v1.1'

average_payload_mass_f9_v1_1 = df.loc[df['Booster_Version'] == 'F9 v1.1', 'PAYLOAD_MASS__KG_'].mean()
average_payload_mass_f9_v1_1
```

### Task 5

##### List the date when the first succesful landing outcome in ground pad was acheived.


_Hint:Use min function_ 


```python
#% select MIN(date) as Date from SPACEXDATASET where mission_outcome == 'Success'
first_successful_ground_landing_date = df['Date'].min()
first_successful_ground_landing_date
```

### Task 6

##### List the names of the boosters which have success in drone ship and have payload mass greater than 4000 but less than 6000


```python
#%sql select booster_version from SPACEXDATASET where (mission_outcome like 'Success') 
#AND (payload_mass__kg_ BETWEEN 4000 AND 6000) AND (landing__outcome like 'Success (drone ship)')
filtered_boosters = df[(df['Mission_Outcome'] == 'Success') &
                       (df['Landing_Outcome'] == 'Success (drone ship)') &
                       (df['PAYLOAD_MASS__KG_'] > 4000) &
                       (df['PAYLOAD_MASS__KG_'] < 6000)]

filtered_boosters
```

### Task 7




##### List the total number of successful and failure mission outcomes


```python
#%sql SELECT mission_outcome, count(*) as Count FROM SPACEXDATASET GROUP by mission_outcome ORDER BY mission_outcome

outcome_counts = df['Mission_Outcome'].value_counts()
outcome_counts
```

### Task 8



##### List the   names of the booster_versions which have carried the maximum payload mass. Use a subquery


```python
#%sql select DISTINCT(BOOSTER_VERSION) from SPACEXTBL where PAYLOAD_MASS__KG_ = (select MAX(PAYLOAD_MASS__KG_) FROM SPACEXDATASET)

max_payload_index = df['PAYLOAD_MASS__KG_'].idxmax()
max_payload_boosters = df.loc[max_payload_index, 'Booster_Version']
max_payload_boosters
```

### Task 9


##### List the records which will display the month names, failure landing_outcomes in drone ship ,booster versions, launch_site for the months in year 2015.

**Note: SQLLite does not support monthnames. So you need to use  substr(Date, 6,2) as month to get the months and substr(Date,0,5)='2015' for year.**


```python
#%sql select MONTHNAME(DATE) as Month, Landing__Outcome, Booster_Version, Launch_Site  from SPACEXDATASET where DATE like '2015%' AND Landing__Outcome like 'Failure (drone ship)'

df['Date'] = pd.to_datetime(df['Date'])

filtered_records = df[(df['Date'].dt.year == 2015)  & (df['Landing_Outcome'] == 'Failure (drone ship)')]

filtered_records['Month'] = filtered_records['Date'].dt.strftime('%B')

result_columns = ['Month', 'Landing_Outcome', 'Booster_Version', 'Launch_Site']
result = filtered_records[result_columns]

result
```

### Task 10




##### Rank the count of landing outcomes (such as Failure (drone ship) or Success (ground pad)) between the date 2010-06-04 and 2017-03-20, in descending order.


```python
#%sql select Landing__Outcome, count(*) as count from SPACEXDATASET where Date >= '2010-06-04' AND Date <= '2017-03-20' GROUP By Landing__Outcome ORDER BY count Desc

df['Date'] = pd.to_datetime(df['Date'])

# Filter rows based on the specified date range
filtered_df = df[(df['Date'] >= '2010-06-04') & (df['Date'] <= '2017-03-20')]


# Rank the count of successful landing outcomes in descending order
ranked_counts = filtered_df['Landing_Outcome'].value_counts(ascending=False)
ranked_counts
```
